```python
from pmdarima import  auto_arima
auto_arima(store_ss['Sales'], seasonal=True  , m = 6, maxiter= 75 )
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX 
model1  =  SARIMAX(store_ss['Sales'], order = (0, 0, 3) ,  seasonal_order = (0, 0, 2, 12) )
results =  model1.fit()
```

```python
pred_rv = results.predict(12)
pred_rv = pred_rv.round()
pred_rv
```

```python
plt.figure(figsize= (8 , 8))
plt.plot(store_ss.Sales , marker ="*" , color ="g" , label ="Actual_Test")
plt.plot(pred_rv , color ="r" , marker ="*" ,label ="Pred_Test")
plt.legend()
```
